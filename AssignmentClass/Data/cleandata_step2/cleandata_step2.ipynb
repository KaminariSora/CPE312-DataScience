```python
import pandas as pd
df = pd.read_csv("clean_combined_file.csv")
df
```

```python
#เพศ
Sex_map = {'ชาย': 'Male', 'หญิง': 'Female'}
df['Sex'] = df['Sex'].map(Sex_map).fillna(df['Sex'])  # เติมค่าเดิมถ้าไม่มีใน map
df
```

```python
#จังหวัด
Sex_map = {'นครนายก': 'Nakhon Nayok', 'นนทบุรี': 'Nonthaburi', 'ปทุมธานี': 'Pathum Thani', 'พระนครศรีอยุธยา': 'Phra Nakhon Si Ayutthaya', 'ลพบุรี': 'Lopburi', 'สระบุรี': 'Saraburi', 'สิงห์บุรี': 'Sing Buri', 'อ่างทอง': 'Ang Thong'}
df['province'] = df['province'].map(Sex_map).fillna(df['province'])  # เติมค่าเดิมถ้าไม่มีใน map

# Map สำหรับเพิ่ม province_id
province_id_map = {
    'Nakhon Nayok': 1, 'Nonthaburi': 2, 'Pathum Thani': 3, 'Phra Nakhon Si Ayutthaya': 4,
    'Lopburi': 5, 'Saraburi': 6, 'Sing Buri': 7, 'Ang Thong': 8
}
df['province_id'] = df['province'].map(province_id_map).fillna(-1)  # ใช้ -1 ถ้าไม่มีใน map

df
```

```python
# แทรกคอลัมน์ province_id ในตำแหน่ง index 1
df.insert(3, 'province_id', df.pop('province_id'))

df
```

```python
# ตั้งค่าทศนิยม 2 ตำแหน่งสำหรับคอลัมน์ 'price' และ 'discount'
df[['MinRain', 'MaxRain', 'AvgRain', 'Tmax', 'Tmin']] = df[['MinRain', 'MaxRain', 'AvgRain', 'Tmax', 'Tmin']].round(2)

df
```

```python
# ลบแถวที่ซ้ำและอัปเดต DataFrame เดิม
df.drop_duplicates(inplace=True)
```

```python
# นับจำนวนแถวที่ซ้ำทั้งหมด
num_duplicates = df.duplicated().sum()
print(num_duplicates)
```

```python
# บันทึกผลลัพธ์ลงไฟล์ CSV ใหม่
df.to_csv("clean_final1.csv", index=False)
df = pd.read_csv('clean_final1.csv')
df
```

```python
df = pd.read_csv('clean_final1.csv')

# ใช้ pivot_table เพื่อจัดกลุ่มตาม year_sick, month_sick, AvgRain, Tmax, Tmin และนับจำนวน Male และ Female ในคอลัมน์ Sex
pivot_table = df.pivot_table(
    index=['province','province_id','year_sick', 'month_sick', 'AvgRain', 'Tmax', 'Tmin'],
    columns='Sex',
    aggfunc='size',
    fill_value=0
)
# รีเซ็ต index และทำให้คอลัมน์เป็นแค่ 1 แถว
pivot_table = pivot_table.reset_index()

# ลบชื่อคอลัมน์ 'Sex' ที่เกิดจาก pivot_table
pivot_table.columns.name = None

# แสดงผลลัพธ์
print("ตารางที่นับจำนวน Male และ Female ตาม year_sick, month_sick, AvgRain, Tmax, และ Tmin:")
pivot_table
```

```python
# เพิ่มคอลัมน์ใหม่สำหรับผลรวมของ Male และ Female
pivot_table['Total'] = pivot_table['Male'] + pivot_table['Female']
pivot_table
```

```python
# จัดลำดับคอลัมน์ใหม่
pivot_table = pivot_table[['Total', 'Male', 'Female', 'AvgRain', 'Tmax', 'Tmin','province' , 'province_id','month_sick','year_sick']]

# บันทึก pivot_table เป็นไฟล์ .csv
pivot_table.to_csv('pivot_output.csv', index=False)

print("บันทึก pivot_table เสร็จสิ้นในไฟล์ 'pivot_output.csv'")
```
