```python
import numpy as np
# กำหนดกลุ่มจังหวัดที่ต้องการใช้ในการคำนวณค่าเฉลี่ย
target_provinces = ['พระนครศรีอยุธยา', 'ลพบุรี', 'สุพรรณบุรี']

# คำนวณค่าเฉลี่ย Tmax สำหรับจังหวัดที่กำหนดในแต่ละ (year_sick, month_sick)
avg_tmax = (
    df[df['province'].isin(target_provinces)]
    .groupby(['year_sick', 'month_sick'])['Tmax']
    .mean()
    .reset_index()
    .rename(columns={'Tmax': 'avg_Tmax'})
)

# รวมค่าเฉลี่ยที่ได้กลับเข้ากับ DataFrame หลัก
df = df.merge(avg_tmax, on=['year_sick', 'month_sick'], how='left')

# เติมค่า Tmax ในจังหวัด "อ่างทอง" ด้วยค่าเฉลี่ยที่คำนวณได้
df.loc[(df['province'] == 'อ่างทอง') & (df['Tmax'].isna()), 'Tmax'] = df['avg_Tmax']

# ลบคอลัมน์ค่าเฉลี่ยออก (ถ้าไม่ต้องการเก็บไว้)
df = df.drop(columns=['avg_Tmax'])

print(df)
```
