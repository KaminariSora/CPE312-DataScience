```python
import pandas as pd
import glob

# ระบุพาธของไฟล์ทั้งหมดที่ต้องการรวม (ในที่นี้ใช้ *.csv เพื่อเลือกไฟล์ CSV ทั้งหมดในโฟลเดอร์)
all_files = glob.glob("*.csv")

# เริ่มจากการอ่านไฟล์แรกเพื่อเป็นฐานข้อมูลเริ่มต้น
combined_df = pd.read_csv(all_files[0])

# วนลูปอ่านและรวมไฟล์ที่เหลือ
for file in all_files[1:]:
    df = pd.read_csv(file)
    # รวมข้อมูลโดยใช้ "จังหวัด", "เดือน" และ "ปี" เป็นตัวเชื่อม
    combined_df = pd.merge(combined_df, df, on=['province', 'month_sick', 'year_sick'], how="left")  # ใช้ outer join เพื่อรักษาข้อมูลทั้งหมด

# บันทึกผลลัพธ์ลงไฟล์ CSV ใหม่
combined_df.to_csv("combined_file.csv", index=False)
```
